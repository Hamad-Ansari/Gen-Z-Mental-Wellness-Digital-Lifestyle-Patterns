## Target Variables Analysis

```python
"""
═══════════════════════════════════════════════════════════════════════════════
🎯 TARGET VARIABLES ANALYSIS
═══════════════════════════════════════════════════════════════════════════════
"""

fig = plt.figure(figsize=(18, 6))
gs = fig.add_gridspec(1, 3, hspace=0.3, wspace=0.3)

# 1. Wellbeing Index Distribution
ax1 = fig.add_subplot(gs[0, 0])
wellbeing_data = df['Wellbeing_Index'].dropna()
n, bins, patches = ax1.hist(wellbeing_data, bins=40, color=PRIMARY_COLOR, 
                            edgecolor='white', alpha=0.8)
ax1.axvline(wellbeing_data.mean(), color='red', linestyle='--', linewidth=2, 
           label=f'Mean: {wellbeing_data.mean():.2f}')
ax1.axvline(wellbeing_data.median(), color='green', linestyle='--', linewidth=2,
           label=f'Median: {wellbeing_data.median():.2f}')
ax1.set_xlabel('Wellbeing Index', fontsize=12, fontweight='bold')
ax1.set_ylabel('Frequency', fontsize=12, fontweight='bold')
ax1.set_title('🌟 Wellbeing Index Distribution\n(Regression Target)', 
             fontsize=14, fontweight='bold', pad=20)
ax1.legend()
ax1.grid(alpha=0.3)

# 2. Burnout Risk Distribution
ax2 = fig.add_subplot(gs[0, 1])
burnout_counts = df['Burnout_Risk'].value_counts()
colors_burnout = {'Low': SUCCESS_COLOR, 'Medium': WARNING_COLOR, 'High': DANGER_COLOR}
bars = ax2.bar(burnout_counts.index, burnout_counts.values, 
              color=[colors_burnout.get(x, PRIMARY_COLOR) for x in burnout_counts.index],
              edgecolor='white', linewidth=2)

# Add value labels
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}\n({height/len(df)*100:.1f}%)',
            ha='center', va='bottom', fontweight='bold', fontsize=11)

ax2.set_xlabel('Burnout Risk Level', fontsize=12, fontweight='bold')
ax2.set_ylabel('Count', fontsize=12, fontweight='bold')
ax2.set_title('🔥 Burnout Risk Distribution\n(Classification Target)', 
             fontsize=14, fontweight='bold', pad=20)
ax2.grid(axis='y', alpha=0.3)

# 3. Wellbeing by Burnout Risk
ax3 = fig.add_subplot(gs[0, 2])
burnout_order = ['Low', 'Medium', 'High']
bp = ax3.boxplot([df[df['Burnout_Risk']==risk]['Wellbeing_Index'].dropna() 
                   for risk in burnout_order],
                  labels=burnout_order,
                  patch_artist=True,
                  notch=True,
                  showmeans=True)

for patch, color in zip(bp['boxes'], [SUCCESS_COLOR, WARNING_COLOR, DANGER_COLOR]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax3.set_xlabel('Burnout Risk', fontsize=12, fontweight='bold')
ax3.set_ylabel('Wellbeing Index', fontsize=12, fontweight='bold')
ax3.set_title('📊 Wellbeing Index by Burnout Risk', 
             fontsize=14, fontweight='bold', pad=20)
ax3.grid(alpha=0.3)

plt.suptitle('🎯 TARGET VARIABLES COMPREHENSIVE ANALYSIS', 
            fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Print Statistics
print("="*80)
print("📊 TARGET VARIABLES STATISTICS")
print("="*80)

print("\n🌟 Wellbeing Index (Regression Target):")
print(f"   • Mean: {wellbeing_data.mean():.2f}")
print(f"   • Median: {wellbeing_data.median():.2f}")
print(f"   • Std Dev: {wellbeing_data.std():.2f}")
print(f"   • Range: [{wellbeing_data.min():.2f}, {wellbeing_data.max():.2f}]")
print(f"   • Skewness: {wellbeing_data.skew():.2f}")

print("\n🔥 Burnout Risk (Classification Target):")
for risk in burnout_order:
    count = (df['Burnout_Risk'] == risk).sum()
    pct = count / len(df) * 100
    print(f"   • {risk}: {count} ({pct:.2f}%)")
```

## Distribution Plots

```python
"""
═══════════════════════════════════════════════════════════════════════════════
📊 FEATURE DISTRIBUTIONS - PSYCHOLOGICAL INDICATORS
═══════════════════════════════════════════════════════════════════════════════
"""

# Psychological Indicators
psych_cols = ['Overthinking_Score', 'Anxiety_Score', 'Mood_Stability_Score',
              'Social_Comparison_Index', 'Sleep_Quality_Score', 
              'Motivation_Level', 'Emotional_Fatigue_Score']

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

for idx, col in enumerate(psych_cols):
    ax = axes[idx]
    data = df[col].dropna()
    
    # Histogram with KDE
    ax.hist(data, bins=30, alpha=0.6, color=CUSTOM_PALETTE[idx % len(CUSTOM_PALETTE)],
           edgecolor='black', linewidth=0.5, density=True, label='Distribution')
    
    # KDE overlay
    from scipy.stats import gaussian_kde
    density = gaussian_kde(data)
    xs = np.linspace(data.min(), data.max(), 200)
    ax.plot(xs, density(xs), color='darkred', linewidth=2, label='KDE')
    
    # Statistics
    mean_val = data.mean()
    median_val = data.median()
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, 
              label=f'Mean: {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2,
              label=f'Median: {median_val:.1f}')
    
    ax.set_xlabel(col.replace('_', ' '), fontsize=10, fontweight='bold')
    ax.set_ylabel('Density', fontsize=10)
    ax.set_title(f'📊 {col.replace("_", " ")}', fontsize=11, fontweight='bold')
    ax.legend(fontsize=8, loc='upper right')
    ax.grid(alpha=0.3)

# Hide unused subplots
for idx in range(len(psych_cols), len(axes)):
    axes[idx].axis('off')

plt.suptitle('🧠 PSYCHOLOGICAL INDICATORS DISTRIBUTION ANALYSIS', 
            fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()
```

## Missing Values Analysis

```python
"""
═══════════════════════════════════════════════════════════════════════════════
🔍 MISSING VALUES ANALYSIS & HANDLING
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*25 + "🔍 MISSING VALUES ANALYSIS")
print("="*80)

# Calculate missing values
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing Count': df.isnull().sum().values,
    'Missing %': (df.isnull().sum().values / len(df) * 100).round(2),
    'Data Type': df.dtypes.values
})

missing_data = missing_data[missing_data['Missing Count'] > 0].sort_values('Missing %', ascending=False)

if len(missing_data) > 0:
    print("\n📊 Columns with Missing Values:")
    display(missing_data.style.background_gradient(subset=['Missing %'], cmap='Reds'))
    
    # Visualize
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Missing counts
    ax1 = axes[0]
    bars = ax1.barh(missing_data['Column'], missing_data['Missing Count'],
                   color=sns.color_palette("Reds_r", len(missing_data)),
                   edgecolor='black', linewidth=1.5)
    ax1.set_xlabel('Missing Count', fontsize=12, fontweight='bold')
    ax1.set_title('📊 Missing Values by Column', fontsize=14, fontweight='bold')
    for i, (bar, val) in enumerate(zip(bars, missing_data['Missing Count'])):
        ax1.text(val + max(missing_data['Missing Count'])*0.01, i,
                f"{val} ({missing_data['Missing %'].iloc[i]:.1f}%)",
                va='center', fontweight='bold')
    ax1.grid(axis='x', alpha=0.3)
    
    # Missing heatmap
    ax2 = axes[1]
    missing_cols = missing_data['Column'].tolist()
    sns.heatmap(df[missing_cols].isnull().T, cbar=True, yticklabels=missing_cols,
               cmap='RdYlGn_r', ax=ax2, cbar_kws={'label': 'Missing'})
    ax2.set_title('🗺️ Missing Values Heatmap', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Sample Index')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("\n✅ No missing values found in the dataset!")

# Handle Missing Values
print("\n" + "="*80)
print(" "*25 + "🔧 HANDLING MISSING VALUES")
print("="*80)

df_clean = df.copy()

# Strategy: Use median for numerical, mode for categorical
numerical_cols_with_missing = df_clean.select_dtypes(include=[np.number]).columns[
    df_clean.select_dtypes(include=[np.number]).isnull().any()
].tolist()

categorical_cols_with_missing = df_clean.select_dtypes(include=['object']).columns[
    df_clean.select_dtypes(include=['object']).isnull().any()
].tolist()

if numerical_cols_with_missing:
    print("\n📌 Imputing Numerical Columns (Median Strategy):")
    for col in numerical_cols_with_missing:
        median_val = df_clean[col].median()
        missing_count = df_clean[col].isnull().sum()
        df_clean[col].fillna(median_val, inplace=True)
        print(f"   ✅ {col}: Filled {missing_count} values with median = {median_val:.2f}")

if categorical_cols_with_missing:
    print("\n📌 Imputing Categorical Columns (Mode Strategy):")
    for col in categorical_cols_with_missing:
        mode_val = df_clean[col].mode()[0]
        missing_count = df_clean[col].isnull().sum()
        df_clean[col].fillna(mode_val, inplace=True)
        print(f"   ✅ {col}: Filled {missing_count} values with mode = '{mode_val}'")

# Verify
print("\n" + "-"*80)
print(f"✅ Missing Values After Imputation: {df_clean.isnull().sum().sum()}")
print("✅ Data cleaning completed successfully!")
```

## Outlier Detection & Treatment

```python
"""
═══════════════════════════════════════════════════════════════════════════════
📊 OUTLIER DETECTION & TREATMENT
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*25 + "📊 OUTLIER DETECTION (IQR METHOD)")
print("="*80)

def detect_outliers_iqr(data, column):
    """Detect outliers using IQR method"""
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers), lower_bound, upper_bound, Q1, Q3

# Analyze outliers
numerical_features = df_clean.select_dtypes(include=[np.number]).columns.tolist()
outlier_summary = []

for col in numerical_features:
    count, lb, ub, q1, q3 = detect_outliers_iqr(df_clean, col)
    outlier_summary.append({
        'Feature': col,
        'Outliers': count,
        'Outlier %': round(count / len(df_clean) * 100, 2),
        'Lower Bound': round(lb, 2),
        'Upper Bound': round(ub, 2),
        'Q1': round(q1, 2),
        'Q3': round(q3, 2)
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier %', ascending=False)

print("\n📊 Outlier Analysis Results:")
display(outlier_df.style.background_gradient(subset=['Outlier %'], cmap='Reds'))

# Visualize outliers with box plots
cols_with_outliers = outlier_df[outlier_df['Outliers'] > 0]['Feature'].tolist()[:12]

if cols_with_outliers:
    fig, axes = plt.subplots(3, 4, figsize=(20, 15))
    axes = axes.flatten()
    
    for idx, col in enumerate(cols_with_outliers):
        ax = axes[idx]
        bp = ax.boxplot(df_clean[col].dropna(), patch_artist=True, notch=True,
                       boxprops=dict(facecolor=CUSTOM_PALETTE[idx % len(CUSTOM_PALETTE)], alpha=0.7),
                       medianprops=dict(color='red', linewidth=2),
                       whiskerprops=dict(color='black', linewidth=1.5),
                       capprops=dict(color='black', linewidth=1.5))
        ax.set_title(f'{col}\n({outlier_df[outlier_df["Feature"]==col]["Outlier %"].values[0]:.1f}% outliers)',
                    fontsize=10, fontweight='bold')
        ax.set_ylabel('Value', fontsize=9)
        ax.grid(alpha=0.3)
    
    # Hide empty subplots
    for idx in range(len(cols_with_outliers), len(axes)):
        axes[idx].axis('off')
    
    plt.suptitle('📊 BOX PLOTS - OUTLIER DETECTION', fontsize=16, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.show()

# Handle outliers using Winsorization (capping)
print("\n" + "="*80)
print(" "*25 + "🔧 OUTLIER TREATMENT (WINSORIZATION)")
print("="*80)

def winsorize_column(data, column, lower_percentile=5, upper_percentile=95):
    """Cap outliers at specified percentiles"""
    lower = data[column].quantile(lower_percentile / 100)
    upper = data[column].quantile(upper_percentile / 100)
    original_outliers = len(data[(data[column] < lower) | (data[column] > upper)])
    data[column] = data[column].clip(lower=lower, upper=upper)
    return original_outliers, lower, upper

# Apply winsorization to features with significant outliers
features_to_winsorize = outlier_df[outlier_df['Outlier %'] > 2]['Feature'].tolist()

print("\n📌 Applying Winsorization (5th-95th percentile capping):")
for col in features_to_winsorize:
    outliers_capped, lower, upper = winsorize_column(df_clean, col)
    print(f"   ✅ {col}: {outliers_capped} outliers capped to [{lower:.2f}, {upper:.2f}]")

print("\n✅ Outlier treatment completed successfully!")
```

## Feature Engineering & Encoding

```python
"""
═══════════════════════════════════════════════════════════════════════════════
🏷️ FEATURE ENGINEERING & ENCODING
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*25 + "🔧 FEATURE ENGINEERING")
print("="*80)

# Create new features
print("\n📌 Creating New Engineered Features:")

# 1. Digital Intensity Score
df_clean['Digital_Intensity'] = (df_clean['Daily_Social_Media_Hours'] + 
                                  df_clean['Screen_Time_Hours'] + 
                                  df_clean['Online_Gaming_Hours']) / 3
print("   ✅ Digital_Intensity: Average of all digital activities")

# 2. Sleep Deficit Indicator
df_clean['Sleep_Deficit'] = (df_clean['Daily_Sleep_Hours'] < 7).astype(int)
print("   ✅ Sleep_Deficit: Binary indicator for inadequate sleep (<7 hours)")

# 3. Mental Load Score
df_clean['Mental_Load'] = (df_clean['Anxiety_Score'] + 
                           df_clean['Overthinking_Score'] + 
                           df_clean['Emotional_Fatigue_Score']) / 3
print("   ✅ Mental_Load: Combined psychological burden score")

# 4. Lifestyle Balance Score
df_clean['Lifestyle_Balance'] = (df_clean['Exercise_Frequency_per_Week'] * 0.5 + 
                                 df_clean['Daily_Sleep_Hours'] * 0.3 - 
                                 df_clean['Caffeine_Intake_Cups'] * 0.2)
print("   ✅ Lifestyle_Balance: Weighted lifestyle indicator")

# 5. Wellbeing Risk Score
df_clean['Wellbeing_Risk'] = (10 - df_clean['Mood_Stability_Score'] + 
                              10 - df_clean['Sleep_Quality_Score'] + 
                              10 - df_clean['Motivation_Level']) / 3
print("   ✅ Wellbeing_Risk: Inverse wellbeing indicator")

# 6. Age Group
df_clean['Age_Group'] = pd.cut(df_clean['Age'], 
                              bins=[17, 20, 23, 27], 
                              labels=['18-20', '21-23', '24-26'])
print("   ✅ Age_Group: Categorical age grouping")

# 7. Screen Time Category
df_clean['Screen_Time_Category'] = pd.cut(df_clean['Screen_Time_Hours'],
                                          bins=[0, 4, 8, 12, 24],
                                          labels=['Low', 'Moderate', 'High', 'Extreme'])
print("   ✅ Screen_Time_Category: Screen time intensity level")

print(f"\n📊 New Dataset Shape: {df_clean.shape}")

# Encode Categorical Variables
print("\n" + "="*80)
print(" "*25 + "🏷️ ENCODING CATEGORICAL VARIABLES")
print("="*80)

# Label Encoding for categorical features
label_encoders = {}
categorical_features = ['Gender', 'Country', 'Student_Working_Status',
                       'Night_Scrolling_Frequency', 'Content_Type_Preference',
                       'Age_Group', 'Screen_Time_Category']

print("\n📌 Applying Label Encoding:")
for col in categorical_features:
    if col in df_clean.columns:
        le = LabelEncoder()
        df_clean[f'{col}_Encoded'] = le.fit_transform(df_clean[col].astype(str))
        label_encoders[col] = le
        print(f"   ✅ {col} → {col}_Encoded (Classes: {len(le.classes_)})")

# Encode target variable
burnout_encoder = LabelEncoder()
df_clean['Burnout_Risk_Encoded'] = burnout_encoder.fit_transform(df_clean['Burnout_Risk'])
print(f"\n📌 Target Variable Encoding:")
print(f"   ✅ Burnout_Risk → Burnout_Risk_Encoded")
print(f"      Mapping: {dict(zip(burnout_encoder.classes_, burnout_encoder.transform(burnout_encoder.classes_)))}")

print("\n✅ Feature engineering and encoding completed successfully!")
```

## Model Preparation

```python
"""
═══════════════════════════════════════════════════════════════════════════════
📋 MODEL PREPARATION - FEATURE SELECTION & SCALING
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*25 + "📋 PREPARING DATA FOR MODELING")
print("="*80)

# Define features to exclude
exclude_cols = ['Wellbeing_Index', 'Burnout_Risk', 'Burnout_Risk_Encoded',
                'Gender', 'Country', 'Student_Working_Status',
                'Night_Scrolling_Frequency', 'Content_Type_Preference',
                'Age_Group', 'Screen_Time_Category']

# Select feature columns
feature_columns = [col for col in df_clean.columns 
                  if col not in exclude_cols and pd.api.types.is_numeric_dtype(df_clean[col])]

print(f"\n📊 Selected Feature Columns ({len(feature_columns)}):")
print("-" * 80)
for i, col in enumerate(feature_columns, 1):
    print(f"   {i:2d}. {col}")

# Create feature matrix
X = df_clean[feature_columns].copy()

# Target variables
y_classification = df_clean['Burnout_Risk_Encoded']  # Classification target
y_regression = df_clean['Wellbeing_Index']  # Regression target

print(f"\n📊 Data Shapes:")
print(f"   • Feature Matrix (X): {X.shape}")
print(f"   • Classification Target: {y_classification.shape}")
print(f"   • Regression Target: {y_regression.shape}")

# Check for any remaining missing values
if X.isnull().sum().sum() > 0:
    print("\n⚠️ Warning: Missing values detected. Filling with median...")
    X = X.fillna(X.median())

# Feature Scaling
print("\n" + "="*80)
print(" "*25 + "⚖️ FEATURE SCALING")
print("="*80)

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)

print("\n✅ Features scaled using StandardScaler (mean=0, std=1)")

# Train-Test Split
print("\n" + "="*80)
print(" "*25 + "✂️ TRAIN-TEST SPLIT")
print("="*80)

# Classification split (stratified)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_scaled, y_classification, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_classification
)

# Regression split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_scaled, y_regression,
    test_size=0.2,
    random_state=42
)

print("\n📊 Classification Split:")
print(f"   • Training Set: {X_train_clf.shape[0]:,} samples ({X_train_clf.shape[0]/len(X)*100:.1f}%)")
print(f"   • Testing Set:  {X_test_clf.shape[0]:,} samples ({X_test_clf.shape[0]/len(X)*100:.1f}%)")

print("\n📊 Regression Split:")
print(f"   • Training Set: {X_train_reg.shape[0]:,} samples ({X_train_reg.shape[0]/len(X)*100:.1f}%)")
print(f"   • Testing Set:  {X_test_reg.shape[0]:,} samples ({X_test_reg.shape[0]/len(X)*100:.1f}%)")

# Class distribution
print("\n📊 Classification Target Distribution:")
print("\n   Training Set:")
train_dist = y_train_clf.value_counts().sort_index()
for label, count in train_dist.items():
    class_name = burnout_encoder.inverse_transform([label])[0]
    print(f"      • {class_name:8s}: {count:,} ({count/len(y_train_clf)*100:.1f}%)")

print("\n   Testing Set:")
test_dist = y_test_clf.value_counts().sort_index()
for label, count in test_dist.items():
    class_name = burnout_encoder.inverse_transform([label])[0]
    print(f"      • {class_name:8s}: {count:,} ({count/len(y_test_clf)*100:.1f}%)")

print("\n✅ Data preparation completed successfully!")
```

## Classification Models Training

```python
"""
═══════════════════════════════════════════════════════════════════════════════
🎯 CLASSIFICATION MODELS - BURNOUT RISK PREDICTION
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*15 + "🎯 TRAINING CLASSIFICATION MODELS - BURNOUT RISK")
print("="*80)

# Define classification models
classification_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=10),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(probability=True, random_state=42),
    'Extra Trees': ExtraTreesClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'AdaBoost': AdaBoostClassifier(random_state=42)
}

# Add advanced models if available
if ADVANCED_MODELS:
    classification_models['XGBoost'] = XGBClassifier(random_state=42, eval_metric='mlogloss', n_jobs=-1)
    classification_models['LightGBM'] = LGBMClassifier(random_state=42, verbose=-1, n_jobs=-1)

# Train and evaluate
clf_results = []
trained_classifiers = {}

print("\n🚀 Training and Evaluating Models...")
print("-" * 80)

for name, model in classification_models.items():
    print(f"\n⏳ Training {name}...", end=' ')
    
    # Train
    model.fit(X_train_clf, y_train_clf)
    
    # Predictions
    y_pred_train = model.predict(X_train_clf)
    y_pred_test = model.predict(X_test_clf)
    
    # Metrics
    train_accuracy = accuracy_score(y_train_clf, y_pred_train)
    test_accuracy = accuracy_score(y_test_clf, y_pred_test)
    precision = precision_score(y_test_clf, y_pred_test, average='weighted', zero_division=0)
    recall = recall_score(y_test_clf, y_pred_test, average='weighted')
    f1 = f1_score(y_test_clf, y_pred_test, average='weighted')
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_scaled, y_classification, cv=5, 
                                scoring='accuracy', n_jobs=-1)
    
    # Store results
    clf_results.append({
        'Model': name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Overfitting': train_accuracy - test_accuracy
    })
    
    trained_classifiers[name] = model
    
    print(f"✅ Accuracy: {test_accuracy:.4f}, F1: {f1:.4f}")

# Results DataFrame
clf_results_df = pd.DataFrame(clf_results).sort_values('F1-Score', ascending=False)

print("\n" + "="*80)
print(" "*20 + "📊 CLASSIFICATION MODELS COMPARISON")
print("="*80)

display(clf_results_df.style.background_gradient(subset=['Test Accuracy', 'F1-Score', 'CV Mean'], 
                                                 cmap='Greens'))

# Identify best model
best_clf_name = clf_results_df.iloc[0]['Model']
best_clf_f1 = clf_results_df.iloc[0]['F1-Score']

print(f"\n🏆 BEST MODEL: {best_clf_name} (F1-Score: {best_clf_f1:.4f})")
```

##  Regression Models Training

```python
"""
═══════════════════════════════════════════════════════════════════════════════
📈 REGRESSION MODELS - WELLBEING INDEX PREDICTION
═══════════════════════════════════════════════════════════════════════════════
"""

print("="*80)
print(" "*15 + "📈 TRAINING REGRESSION MODELS - WELLBEING INDEX")
print("="*80)

# Define regression models
from sklearn.linear_model import LinearRegression

regression_models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0, random_state=42),
    'Lasso Regression': Lasso(alpha=0.01, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    
    
    'SVR': SVR()
}

# Add advanced models if available
if ADVANCED_MODELS:
    regression_models['XGBoost'] = XGBRegressor(random_state=42, n_jobs=-1)
    regression_models['LightGBM'] = LGBMRegressor(random_state=42, verbose=-1, n_jobs=-1)

# Train and evaluate
reg_results = []
trained_regressors = {}

print("\n🚀 Training and Evaluating Models...")
print("-" * 80)

for name, model in regression_models.items():
    print(f"\n⏳ Training {name}...", end=' ')
    
    # Train
    model.fit(X_train_reg, y_train_reg)
    
    # Predictions
    y_pred_train = model.predict(X_train_reg)
    y_pred_test = model.predict(X_test_reg)
    
    # Metrics
    train_r2 = r2_score(y_train_reg, y_pred_train)
    test_r2 = r2_score(y_test_reg, y_pred_test)
    mse = mean_squared_error(y_test_reg, y_pred_test)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_reg, y_pred_test)
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_scaled, y_regression, cv=5, 
                                scoring='r2', n_jobs=-1)
    
    # Store results
    reg_results.append({
        'Model': name,
        'Train R²': train_r2,
        'Test R²': test_r2,
        'RMSE': rmse,
        'MAE': mae,
        'CV Mean R²': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Overfitting': train_r2 - test_r2
    })
    
    trained_regressors[name] = model
    
    print(f"✅ R²: {test_r2:.4f}, RMSE: {rmse:.4f}")

# Results DataFrame
reg_results_df = pd.DataFrame(reg_results).sort_values('Test R²', ascending=False)

print("\n" + "="*80)
print(" "*20 + "📊 REGRESSION MODELS COMPARISON")
print("="*80)

display(reg_results_df.style.background_gradient(subset=['Test R²', 'CV Mean R²'], 
                                                 cmap='Greens').background_gradient(
                                                 subset=['RMSE', 'MAE'], cmap='Reds_r'))

# Identify best model
best_reg_name = reg_results_df.iloc[0]['Model']
best_reg_r2 = reg_results_df.iloc[0]['Test R²']

print(f"\n🏆 BEST MODEL: {best_reg_name} (R²: {best_reg_r2:.4f})")
```
